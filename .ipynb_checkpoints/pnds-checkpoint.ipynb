### DataFrames

```python
dates = pd.date_range("20211009", periods=6)
df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list("BTJX"))

df
```

```python
df2 = pd.DataFrame({
        "A": 1.0,
        "B": pd.Timestamp("20130102"),
        "C": pd.Series(1, index=list(range(4)), dtype="float32"),
        "D": np.array([3] * 4, dtype="int32"),
        "E": pd.Categorical(["test", "train", "test", "train"]),
        "F": "foo",
    })
df2
```
